## 4. 입력값이 소수(Prime Number)인지 아닌지 간별하는 함수를 만들고 테스트 하시오. 


```python
def isPrimeNumber(x):
  for i in range(2, x):
    if x %i==0:
      return False
  return True

print(isPrimeNumber(5))
print(isPrimeNumber(6))
```

```python
def isPrime(n):

  if n == 2:
    return True
  for i in range(2,n/2):
    if input(int(n/i))==n/i:
      return False
    else:
      return True
  print(isPrime(10))
```
