### 함수 
* def는 함수를 만들 때 사용하는 예약어

```python
def add(a,b):
  return a+b

add(5,6)
```

```python
def add(a,b=2):
  return a+b

add(5)
```

```python
def add(a,b=2):
  '''
  더하기 함수 
    usage:
       add(2)  # 4
  '''
  return a+b

add(5)
```
